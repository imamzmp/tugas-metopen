```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from google.colab.patches import cv2_imshow  # Hanya untuk Google Colab
import warnings
warnings.filterwarnings('ignore')
```

```python
from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    image_path = filename
    break

# Baca dan tampilkan gambar
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('Gambar Asli')
plt.axis('off')
plt.show()
```

```python
pixel_vals = img_rgb.reshape((-1, 3))
pixel_vals = np.float32(pixel_vals)

# Terapkan k-means
k = 3  # jumlah klaster warna
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(pixel_vals)
centers = np.uint8(kmeans.cluster_centers_)
segmented_data = centers[labels.flatten()]
segmented_image = segmented_data.reshape(img_rgb.shape)

# Tampilkan hasil segmentasi
plt.imshow(segmented_image)
plt.title('Hasil Segmentasi Warna ({} Klaster)'.format(k))
plt.axis('off')
plt.show()
```

```python
output_path = 'hasil_segmentasi.png'
cv2.imwrite(output_path, cv2.cvtColor(segmented_image, cv2.COLOR_RGB2BGR))
print('Gambar hasil segmentasi telah disimpan sebagai:', output_path)
```
